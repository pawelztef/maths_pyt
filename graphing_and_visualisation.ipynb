```python
x = [-4, 5, 2, 4, 5, 6, 3]
y = [0, 3, 4, 5, 2, -3, 4]

for i in range(0, len(x)):
    plt.plot(x[i], y[i], 'o', label=f'point {1+i}')

plt.axis('square')
plt.grid()
plt.legend(loc=2, numpoints=1)

plt.show()
```

##### Excercise

$ 
y = x^2 - 3x \\ 
x \in \{-10, -9, \text{ ..., } 10, 11\} 
$

```python
import sympy as sym
import matplotlib.pyplot as plt
from IPython.display import display, Math

x = sym.symbols('x')
fn = x**2 - 3*x

for fn_x in range(-9, 10):
    fn_y = fn.subs(x, fn_x)
    plt.plot(fn_x, fn_y, 'o')

plt.xlabel('x')
plt.ylabel('f(x) = %s' %sym.latex(fn))
plt.show()
    
```
