##### Excercise

$ 
y = x^2 - 3x \\ 
x \in \{-10, -9, \text{ ..., } 10, 11\} 
$

```python
import sympy as sym
import matplotlib.pyplot as plt
from IPython.display import display, Math

x = sym.symbols('x')
fn = x**2 - 3*x

for fn_x in range(-9, 10):
    fn_y = fn.subs(x, fn_x)
    plt.plot(fn_x, fn_y, 'o')

plt.xlabel('x')
plt.ylabel('f(x) = %s' %sym.latex(fn))
plt.show()
    
```

#### Lines in slope-intercept form

$ 
y = mx + b \\
\text{ m = "slope"} \\
\text{ b = "intercept" }
$



```python
import numpy as np
import matplotlib.pyplot as plt

# y = mx + b

x = [-5, 5]
m = 2
b = 1

# y = [340, 2345]
# for i in range(0, len(x)):
#    y[i] = m*x[i] + b

y = m*np.array(x) + b

print(x, y)
plt.plot(x, y)
plt.grid()

plt.plot(x, y, label='y=%sx+%s' %(m, b))
plt.axis('square')
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(), [0, 0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')
print(axis.get_xlim())
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = [-5, 5]

data = [(.7, -2), (-1.25, .75)]

for d in data:
    y = d[0]*np.array(x)+d[1]
    plt.plot(x, y, label='y=%sx+%s' %(d[0], d[1]))
    
plt.xlim(x)
plt.ylim(x)

plt.axis('square')
axis = plt.gca()
print(axis.get_xlim())
print(axis.get_ylim())
plt.plot(axis.get_xlim(), [0, 0], 'k--')
plt.plot([0, 0], axis.get_ylim(), 'k--')

plt.legend()
plt.grid()
plt.show()
```

##### Excercise
Plot with Sympy

$
y = \frac{a}{x^{2}-a}
$

```python
import sympy as sym
import sympy.plotting.plot as symplot

x, y, a = sym.symbols('x, y, a')

y = a /(x**2 - a)

p = symplot(y.subs(a, 1), show=False)
p[0].label = '$%s$' % sym.latex(y.subs(a, 1))
p.xlim = [-5, 5]

for ia in range(2, 5):
    p_temp = symplot(y.subs(a, ia), show=False)
    p.extend(p_temp)
    p[ia-1].label = '$%s$' % sym.latex(y.subs(a, ia))
    
p.ylim = [-10, 10]
p.legend = True
p.show()
```

##### Exercise
Implement algorithm
$
c_{i, j} = -1^{i+j}\\
$

and represent with mathplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

N = 10
mtx = np.zeros((N, N))

for i in range(0, N):
    for j in range(0, N):
        mtx[i, j] = (-1)**(i+j)

plt.imshow(mtx, interpolation='none')
plt.set_cmap('gray')
plt.tick_params(labelleft=False, labelright=False, labelbottom=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Polygon

y = np.array([[1,1], [2,3], [3,1]])
p = Polygon(y, facecolor='m', alpha=.3)
y1 = np.array([[2,2], [2.5,4], [3.5,1]])
p1 = Polygon(y1, facecolor='w', alpha=.3, edgecolor='b')

fig, ax = plt.subplots()
ax.add_patch(p)
ax.add_patch(p1)

ax.set_ylim([0,4])
ax.set_xlim([0,4])
plt.show()
```
