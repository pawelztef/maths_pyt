##### Random numbers

typical distribution of random numbers
* normal
* uniform


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use('classic')
```

```python
nums = np.random.rand(1000)

minv = 2
maxv = 17
nums = nums*(maxv-minv) + minv

plt.plot(nums, 's')
plt.show()

plt.hist(nums)
plt.show()
```

```python
nums = np.random.randn(1000) # normal distribution
plt.plot(nums, 's', alpha=.5)
plt.show()
plt.hist(nums)
plt.show()
```

##### Exercise
Generate a normal distribution with:
* mean = 15
* standart deviation = 4.3

```python
mean = 15
deviation = 4.3
nums = np.random.randn(1000)
nums = nums - np.mean(nums)
nums = nums/np.std(nums)
nums = nums*deviation + mean

plt.subplot(121)
plt.plot(nums, 's')
plt.subplot(122)
plt.hist(nums)
plt.show()

print('This distribution has a mean of %s and standard deviation of %s' %(np.mean(nums), np.std(nums)))
```

##### Converting between radians and degrees
$ 2\pi = 360^o $

$ \text{rad.} = \dfrac{360}{2\pi} = \dfrac{180}{\pi} $

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
deg = 90
rad = deg*np.pi/180
rad = rad%(2*np.pi)

print('%g degrees is %g radians.' %(deg, rad))

rad = 4*np.pi
deg = (180*rad/np.pi)%340
print('%g radians is %g degrees.' %(rad, deg))

# numpy
# rad = np.deg2rad(deg)
# deg = np.rad2deg(rad)
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = 2
b = 3

c = np.sqrt(a**2 + b**2)

plt.plot([0, a],  [0, 0], 'k', linewidth=2)
plt.plot([0, 0],  [0, b], 'k', linewidth=2)
plt.plot([0, a],  [b, 0], 'k', linewidth=2)
plt.plot([.3, .3], [0, .3], 'k')
plt.plot([0, .3], [.3, .3], 'k')

plt.text(a/3, -.35, 'a=' + str(a), fontsize=15)
plt.text(.1, b/3, 'b=' + str(b), fontsize=15)
plt.text(a/2+.2, b/2, 'c=' + str(c), fontsize=15)

plt.axis('square')
axlim = np.max((a, b)) + 0.5
plt.axis([-.5, axlim, -.5, axlim])

plt.show()
```

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt

thsym = sym.pi*1/8
theta = float(thsym)

# plot the unit circle
x = np.linspace(0, 2*np.pi, 200)
plt.plot(np.cos(x), np.sin(x), color='gray')
plt.plot([-1.1, 1.1], [0, 0], '--', color='gray')
plt.plot([0, 0], [-1.1, 1.1], '--', color='gray')

# plot the arc
plt.plot(np.cos(np.linspace(0, theta)), np.sin(np.linspace(0, theta)), 'k', linewidth=3)
plt.plot([0, np.cos(theta)],[0, np.sin(theta)] ,'k:')
plt.plot(np.cos(theta), np.sin(theta), 'ko')


# plot the separete sin/cos components
plt.plot([0,0], [0, np.sin(theta)], 'r')
plt.plot([0, np.cos(theta)], [np.sin(theta), np.sin(theta)] , 'r:')
plt.plot([0, np.cos(theta)], [0,0], 'b')
plt.plot([np.cos(theta), np.cos(theta)], [0, np.sin(theta)] , 'b:')


plt.title('$\\theta = %s$' %sym.latex(thsym))
plt.gca().set_aspect('equal')
plt.xlim([-1.2, 1.2])
plt.ylim([-1.2, 1.2])
plt.show()
```
