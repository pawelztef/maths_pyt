##### Parabolas

$
4p(y-k)=(x-h)^2 \\
y=\frac{1}{4p}(x-h)^2+k \\
$

```python
from matplotlib import style
style.use('classic')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# parameters
a = 1 
h = 1
k = -2
n = 100

# x-axis points to evalueate the function
x = np.linspace(h-5, h+5, n)

y = a*(x-h)**2 + k
plt.plot(x, y)
plt.plot(y, x)
plt.grid()
plt.axis('square')
plt.show()
```

##### Excercises
Vertex: $(h,k)$

Focus: $(h, k+p)$

Directrix: $y=k-p$

```python

# parameters
a = 1 # 1/4p = a, 1 = a4p, 1/4a
p = 1/4*a
h = 1
k = -2
n = 100

# x-axis points to evalueate the function
x = np.linspace(h-5, h+5, n)

y = a*(x-h)**2 + k

# plot the parabola
plt.plot(x, y, label='Parabola')

# plot the vertex
plt.plot(h, k, 'ro', label='Vertex')

# plot the focus 
plt.plot(h, k+p, 'go', label='Focus')

# plot the directirx
plt.plot(x[0, -1, k-p, 'rs', label='Directix')

plt.grid()
plt.show()
```
