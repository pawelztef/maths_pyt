##### Parabolas

$
4p(y-k)=(x-h)^2 \\
y=\frac{1}{4p}(x-h)^2+k \\
$

```python
from matplotlib import style
style.use('classic')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# parameters
a = 1 
h = 1
k = -2
n = 100

# x-axis points to evalueate the function
x = np.linspace(h-5, h+5, n)

y = a*(x-h)**2 + k
plt.plot(x, y)
plt.plot(y, x)
plt.grid()
plt.axis('square')
plt.show()
```

##### Excercises
Vertex: $(h,k)$

Focus: $(h, k+p)$

Directrix: $y=k-p$

```python

# parameters
a = 1 # 1/4p = a, 1 = a4p, 1/4a
p = 1/4*a
h = 1
k = -2
n = 100

# x-axis points to evalueate the function
x = np.linspace(h-1, h+1, n)

y = a*(x-h)**2 + k

# plot the parabola
plt.plot(x, y, label='Parabola')

# plot the vertex
plt.plot(h, k, 'ro', label='Vertex')

# plot the focus 
plt.plot(h, k+p, 'go', label='Focus')

# plot the directirx
d = k-p
plt.plot(x[[0, -1]], [d, d], label='Directix')

plt.grid()
plt.show()
```

##### Creating contours from meshes in Python

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, Y = np.meshgrid(range(0, 10), range(0, 16))
print(X)
print(Y)
```

```python
plt.subplot(121)
plt.pcolormesh(X, edgecolor='k', linewidth=.1)
plt.subplot(122)
plt.pcolormesh(Y, edgecolor='k', linewidth=.1)
plt.show()
```

```python
x = np.linspace(-np.pi, 2*np.pi, 40)
y = np.linspace(0, 4*np.pi, 72)
X, Y = np.meshgrid(x,y)
Fxy = np.cos(X) + np.sin(Y)

#plt.imshow(Fxy)
plt.pcolormesh(Fxy)
plt.show()
```

```python
x = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, x)
s = 1

gauss2d = np.exp( -(X**2+Y**2)/s )
plt.imshow(gauss2d)
plt.axis('off')
plt.show()
```

##### Graphing circles

```python
h = 2
k = -3
r = 3

axlim = r + np.max((abs(h), abs(k)))
x = np.linspace(-axlim, axlim, 100)
X, Y = np.meshgrid(x, x)

Fxy = (X-h)**2 + (Y-k)**2 - r**2
plt.contour(X, Y, Fxy, 0)
plt.plot([-axlim, axlim], [0, 0], 'k--')
plt.plot([0, 0], [-axlim, axlim], 'k--')
plt.axis('square')
plt.show()
```
